```python
if GF.shape == ML.shape:
    
    same = (GF.values == ML.values).all()
    
    if same:
        
        print("두 데이터셋 동일합니다.")
    
    else:
        
        print("두 데이터셋은 동일하지 않습니다.")
        
else:
    
    print("두 데이터셋의 크기가 다릅니다.")
```

```python
different_rows = []

for index, row1 in GF.iterrows():
    
    row2 = ML.iloc[index]
    
    if not row1.equals(row2):
        
        different_rows.append(index)

different_columns = []

for column in GF.columns:
    
    if not (GF[column] == ML[column]).all():
        
        different_columns.append(column)

if different_rows:
    
    print("다른 행:")
    print(different_rows)
    
if different_columns:
    
    print("다른 열:")
    print(different_columns)
```

```python
different_flow_bytes_rows = []

for index, (value_GF, value_ML) in enumerate(zip(GF['Flow Bytes/s'], ML['Flow Bytes/s'])):
    
    if value_GF != value_ML:
        
        different_flow_bytes_rows.append(index)

if different_flow_bytes_rows:
    
    print("다른 'Flow Bytes/s' 열의 행:")
    print(different_flow_bytes_rows)
    print("GF의 값:")
    print(GF.loc[different_flow_bytes_rows, 'Flow Bytes/s'])
    print("ML의 값:")
    print(ML.loc[different_flow_bytes_rows, 'Flow Bytes/s'])
    
else:
    
    print("'Flow Bytes/s' 열의 값이 모두 동일합니다.")
```
